```sage
import os, json
from pathlib import Path
from IPython.display import Image, display

# Detect format during Quarto render
target = None
info_path = os.environ.get("QUARTO_EXECUTE_INFO")

if info_path:
    with open(info_path, "r", encoding="utf-8") as f:
        info = json.load(f)
    target = info["format"]["identifier"]["target-format"]   # e.g. "html", "pdf"
else:
    # Not running under `quarto render` (e.g. you executed interactively in Jupyter/VS Code)
    target = "html"

# Make output paths stable relative to the notebook file (works under Quarto render)
doc_dir = Path(info["document-path"]).parent if info_path else Path.cwd()
img_path = doc_dir / "plot3d-fallback.png"

p = plot3d(sin(x*y), (x, -3, 3), (y, -3, 3))

if target == "html":
    p  # interactive
else:
    p.save(str(img_path), viewer="tachyon")
    display(Image(filename=str(img_path)))
```
