```python
import os, json
from pathlib import Path
from IPython.display import Image, display

def quarto_target_format():
    info_path = os.environ.get("QUARTO_EXECUTE_INFO")
    if not info_path:
        return None  # running interactively, not via `quarto render`
    with open(info_path, "r", encoding="utf-8") as f:
        info = json.load(f)
    return info["format"]["identifier"]["target-format"], info


info_path = os.environ.get("QUARTO_EXECUTE_INFO")
print(info_path)
target, info = quarto_target_format() or ("html", None)
# print(target, info)
# # choose a stable place to write the png:
# # if Quarto is rendering, use the notebook's directory (from execute info)
# doc_dir = Path(info["document-path"]).parent if info else Path.cwd()
# png = doc_dir / "plot3d-fallback2.png"

g = sin(x*y)
p = plot3d(g, (x, -3, 3), (y, -3, 3))

# if target == "html":
#     p  # interactive
# else:
#     p.save(str(png), viewer="tachyon")
#     display(Image(filename=str(png)))

p.save("plot3d-interactive.html")
```
